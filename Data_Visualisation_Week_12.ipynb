```python
# ðŸ“Š Hands-On Visualization Activity
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ“š Step 1: Load the Dataset
# Monthly Sales Data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June',
              'July', 'August', 'September', 'October', 'November', 'December'],
    'Product A Sales': [120, 130, 150, 170, 160, 180, 200, 190, 210, 220, 230, 240],
    'Product B Sales': [95, 100, 110, 120, 115, 125, 135, 130, 140, 145, 150, 155],
    'Product C Sales': [60, 65, 70, 80, 85, 90, 95, 100, 110, 115, 120, 125]
}

df = pd.DataFrame(data)

# Show the data
df

# Reshape the DataFrame to long format for easier plotting with Seaborn (if not already done)
# This is added here to ensure df_melted is available for subsequent plots
# and to make sure the data preparation is near its usage.
df_melted = df.melt(id_vars=['Month'], var_name='Product', value_name='Sales')
```

```python
# ðŸ› ï¸ Step 3: Your Task
# ðŸ‘‰ TASK: Try changing the above chart to a different type:
#    - Bar chart
#    - Scatter plot
#    - Area chart (stacked if possible)
#    - Heatmap
#    - Etc.

# Example 1: Bar Chart (try modifying this!)
plt.figure(figsize=(12,6))
sns.barplot(x='Month', y='Product A Sales', data=df, color='skyblue')
plt.title('Product A Sales Over Months (Bar Chart)')
plt.xticks(rotation=45)
plt.ylabel('Sales Units')
plt.show()

# ðŸ‘‰ Your Turn Below:
# 1. Try plotting Product B Sales instead.
# 2. Try a scatterplot for Product C Sales.
# 3. Try plotting multiple products together.

# 1. Bar Chart for Product B Sales
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Product B Sales', data=df, color='lightcoral')
plt.title('Product B Sales Over Months (Bar Chart)')
plt.xticks(rotation=45)
plt.ylabel('Sales Units')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Scatter Plot for Product C Sales
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Month', y='Product C Sales', data=df, hue='Month', s=100, palette='viridis')
plt.title('Product C Sales Over Months (Scatter Plot)')
plt.xticks(rotation=45)
plt.ylabel('Sales Units')
plt.grid(True)
plt.show()

# 3. Line Plot for Multiple Products Together
# Reshape the DataFrame to long format for easier plotting with Seaborn
df_melted = df.melt(id_vars=['Month'], var_name='Product', value_name='Sales')

plt.figure(figsize=(14, 7))
sns.lineplot(x='Month', y='Sales', hue='Product', data=df_melted, marker='o', lw=2)
plt.title('Monthly Sales for Products A, B, and C')
plt.xticks(rotation=45)
plt.ylabel('Sales Units')
plt.grid(True)
plt.legend(title='Product')
plt.tight_layout()
plt.show()
```

```python
# 4. Stacked Area Chart for Total Sales Over Time
plt.figure(figsize=(14, 7))

# Extract sales data for each product
product_a_sales = df['Product A Sales']
product_b_sales = df['Product B Sales']
product_c_sales = df['Product C Sales']

plt.stackplot(df['Month'], product_a_sales, product_b_sales, product_c_sales,
              labels=['Product A Sales', 'Product B Sales', 'Product C Sales'], alpha=0.8)

plt.title('Stacked Area Chart of Monthly Sales for Products A, B, and C')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# 5. Heatmap for Overview of All Monthly Sales Data

# Prepare data for heatmap: set Month as index and select only sales columns
heatmap_data = df.set_index('Month')[['Product A Sales', 'Product B Sales', 'Product C Sales']]

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
plt.title('Monthly Sales Overview (Heatmap)')
plt.ylabel('Month')
plt.xlabel('Product')
plt.tight_layout()
plt.show()
```
