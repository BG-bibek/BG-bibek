```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['BMI'], kde=True)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
## Generate boxplot
import numpy as np
import matplotlib.pyplot as plt
# Show the quartile (25%, 50%, 75%) informaiton of the 'age' column
data = df
print(np.percentile(data.age, [25, 50, 75]))
# Generate the boxplot of 'age' column in the dataframe
plt.boxplot(data.age)
# Label the axes
plt.ylabel('age')
plt.title('Box plot for the age column')
# Show the plot
plt.show()
```

```python
# Calculate the variance of 'age' column in the dataframe
print("Variance = ", np.var(data.age))
# Calculate the standard deviation of use car 'price'
print("Standard deviation = ", np.std(data.age))
```

```python
## Generate histogram
import matplotlib.pyplot as plt
# plot the histogram for the 'age' column of the dataframe
plt.hist(data.age,
    bins=20,
    density=True,
color='blue')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram", fontsize=15)
plt.show()
```

```python
## Generate scatter plot
import matplotlib.pyplot as plt
# Generate a scatter plot between the 'age' and 'BMI' columns
plt.scatter(data.age, data.BMI,
color='tomato',
marker='x',
alpha=0.5)
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()
```
