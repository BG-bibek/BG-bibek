# Python programming exercise


Now you have a chance to work on a Python programming exercise. This exercise is to simulate
the random walk:
1. Your throw a dice 100 times (the dice has 6 displaying numbers, ie. 1, 2, 3, 4, 5, 6). In each
dice throw:
• if dice number shown is 1 or 2, you will walk one step backward;
• if dice number shown is 3, 4, or 5, you will walk one step forward;
• if dice number shown is 6, you will throw the dice again, and walk forward the resulting
number of steps shown in that dice throw.
2. The intial step count is 0. And in any cases, step count must not be less than 0.
3. Visualize the random walk process.
4. Simulate the above process for 200 times, and visualize the final steps achieved in each
simulation using histogram.
5. Calculate the probability that the final steps achieved is equal or greater than 60 steps.

```python
import random
import matplotlib.pyplot as plt

def simulate_random_walk():
    step_count = 0
    for _ in range(100):
        dice_throw = random.randint(1, 6)
        if dice_throw in [1, 2]:
            step_count = max(0, step_count - 1)  # Walk backward, but not less than 0
        elif dice_throw in [3, 4, 5]:
            step_count += 1  # Walk forward
        else:  # dice_throw == 6
            extra_steps = random.randint(1, 6)  # Throw the dice again
            step_count += extra_steps  # Walk forward the resulting number of steps
    return step_count

# Simulate the random walk process for 200 times and visualize the results

final_steps = [simulate_random_walk() for _ in range(200)]
plt.hist(final_steps, bins=20)
plt.xlabel('Final Steps')
plt.ylabel('Frequency')
plt.title('Distribution of Final Steps in Random Walk')
plt.show()

# Calculate the probability that the final steps achieved is equal or greater than 60 steps
prob = sum(1 for steps in final_steps if steps >= 60) / len(final_steps)
print(f"Probability of achieving 60 or more steps: {prob:.2f}")
```
