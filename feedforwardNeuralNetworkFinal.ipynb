```python
import numpy as np
```

```python
# -----------------------------
# 1. DATA (AND logic)
# -----------------------------
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

Y = np.array([[0, 0, 0, 1]])

# transpose
X_train = X.T
Y_train = Y
```

```python
# -----------------------------
# 2. INITIALIZE PARAMETERS
# -----------------------------
def initialize_parameters(n_features):
    params = {
        "W": np.zeros((1, n_features)),
        "b": np.zeros((1, 1))
    }
    return params
```

```python

# -----------------------------
# 3. ACTIVATION FUNCTION
# -----------------------------
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

# -----------------------------
# 4. FORWARD PROPAGATION
# -----------------------------
def feed_forward(X, params):
    Z = np.dot(params["W"], X) + params["b"]
    A = sigmoid(Z)
    cache = {"Z": Z, "A": A}
    return cache

# -----------------------------
# 5. COST FUNCTION
# -----------------------------
def compute_cost(Y, A):
    m = Y.shape[1]
    cost = (1/m) * np.sum(-Y*np.log(A) - (1-Y)*np.log(1-A))
    return cost

# -----------------------------
# 6. BACKPROPAGATION
# -----------------------------
def back_propagate(X, Y, cache):
    m = X.shape[1]
    A = cache["A"]

    dZ = (1/m) * (A - Y)
    dW = np.dot(dZ, X.T)
    db = np.sum(dZ, axis=1, keepdims=True)

    grads = {"dW": dW, "db": db}
    return grads

# -----------------------------
# 7. UPDATE PARAMETERS
# -----------------------------
def update(params, grads, lr):
    params["W"] = params["W"] - lr * grads["dW"]
    params["b"] = params["b"] - lr * grads["db"]
    return params

# -----------------------------
# 8. TRAINING LOOP
# -----------------------------
def train(X, Y, epochs=1000, lr=0.1):
    params = initialize_parameters(X.shape[0])

    for i in range(epochs):
        cache = feed_forward(X, params)
        cost = compute_cost(Y, cache["A"])
        grads = back_propagate(X, Y, cache)
        params = update(params, grads, lr)

        if i % 100 == 0:
            print(f"Epoch {i}, Cost: {cost:.4f}")

    return params
```

```python
# -----------------------------
# 9. PREDICTION
# -----------------------------
def predict(X, params):
    cache = feed_forward(X, params)
    probs = cache["A"]
    return (probs > 0.5).astype(int)
```

```python
# -----------------------------
# RUN
# -----------------------------
params = train(X_train, Y_train, epochs=1000, lr=0.5)

preds = predict(X_train, params)

print("\nPredictions:")
print(preds)

print("\nActual:")
print(Y_train)
```
